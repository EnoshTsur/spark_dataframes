```python
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName("basics").getOrCreate()
```

```python
df = spark.read.csv("Diabetes Missing Data.csv", inferSchema=True, header=True)
```

```python
df.show()
```

```python
df.na.drop().show()
```

```python
df.na.drop(thresh=8).show()
```

```python
df.na.drop(thresh=7).show()
```

```python
df.na.drop(subset=['Diastolic_BP']).show()
```

```python
df.na.drop(subset=['Diastolic_BP', 'Skin_Fold']).show()
```

```python
df.na.fill(-9999, subset=["Serum_Insulin"]).show()
```
